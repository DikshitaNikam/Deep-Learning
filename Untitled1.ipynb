## Binary Classification With Keras

### Step 1 - Data ingestion

```python
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/train_loan.csv"
print(url1)
```

```python
import pandas as pd
df = pd.read_csv(url1)
df.head()
```

0 - Loan Rejected
1 - Loan Approved

###  step 3 - separate X and Y (loan_status)
drop id column as it is just a serial

```python
X = df.drop(columns=["id", "loan_status"])
Y = df[["loan_status"]]
```

```python
Y.value_counts()
```

### Stwp 3 - Apply Preprocessing on x

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute  import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
)
```

```python
pre = ColumnTransformer (
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
pre
```

### Step 5 - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

### Step 6 - Model Building

image.png

image.png

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
model = Sequential(
    [
        Input(shape = (xtrain.shape[1],)),
        Dense(32, activation="relu"),
        Dropout(0.5),
        Dense(16, activation="relu"),
        Dropout(0.4),
        Dense(1, activation="sigmoid")
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
es = EarlyStopping(patience=10, restore_best_weights=True)
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=200, callbacks=[es])
```

### Plot the learninng curve

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(hist.history["loss"], label="Training Loss")
plt.plot(hist.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Binary Cross Entropy Loss")
plt.title("Learning Curve")
plt.legend()
plt.show()

```

### Trsining is done properlt

### Step 8 - Out of sample prediction

```python
url2 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/test_loan.csv"
print(url2)
```

```python
xnew = pd.read_csv(url2)
xnew.head()
```

```python
xnew_pre = pre.transform(xnew)
xnew_pre.head()
```

```python

probs = model.predict(xnew_pre)
probs[0:5]
```

```python
preds = [1 if p >= 0.5 else 0 for p in probs]
preds[0:5]
```

```python
res = xnew[["id"]]
res["loan_status"] = preds
res["probability"] = probs
```

```python
res = res.round(4)
res
```

```python
res["loan_status"].value_counts()
```

```python
res["loan_status"].value_counts().plot(kind="bar")
```

```python
res.to_csv("loan_res.csv", index=False)
```
